# Business Objectives

Analyze Trade Performance Over Time

```python
export_data = df[df['INDICATOR_LABEL'].str.contains('exports')]
import_data = df[df['INDICATOR_LABEL'].str.contains('imports')]


exports_by_year = export_data.groupby('TIME_PERIOD')['OBS_VALUE'].sum()
imports_by_year = import_data.groupby('TIME_PERIOD')['OBS_VALUE'].sum()

# Create a single plot to show both trends
plt.figure(figsize=(12, 6))
plt.plot(exports_by_year.index, exports_by_year.values, label='Exports')
plt.plot(imports_by_year.index, imports_by_year.values, label='Imports')
plt.title('Trade Performance Over Time (Exports vs. Imports)')
plt.xlabel('Year')
plt.ylabel('Total Trade Value (in millions)')
plt.legend()
plt.grid(True)
plt.show()
```

- The graph shows Exports (blue line) and Imports (orange line) from 2010 to 2023.
- Both exports and imports have been growing steadily over the years.
- Exports have mostly stayed higher than imports, meaning the country is generally exporting more than it imports.
- Around 2015, both exports and imports dipped slightly but picked up again afterward.
- From 2020 onward, exports grew much faster, especially between 2020–2021, creating a bigger gap between exports and imports.
- By 2023, exports are close to 9 million units, while imports are about 7.5 million units, showing a healthy trade surplus.

Identify Top Contributing Countries/Regions

```python
# Get total exports by country
total_exports_by_country = df[df['INDICATOR_LABEL'].str.contains('exports')].groupby('REF_AREA_LABEL')['OBS_VALUE'].sum().sort_values(ascending=False).head(10)

# Get total imports by country
total_imports_by_country = df[df['INDICATOR_LABEL'].str.contains('imports')].groupby('REF_AREA_LABEL')['OBS_VALUE'].sum().sort_values(ascending=False).head(10)


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))

# Bar chart for top exporters
sns.barplot(
    x=total_exports_by_country.values,
    y=total_exports_by_country.index,
    ax=ax1,
    color="skyblue" 
)
ax1.set_title('Top 10 Global Exporters (Total Value)')
ax1.set_xlabel('Total Export Value (in millions)')
ax1.set_ylabel('Country/Region')

# Bar chart for top importers
sns.barplot(
    x=total_imports_by_country.values,
    y=total_imports_by_country.index,
    ax=ax2,
    color="salmon" 
)
ax2.set_title('Top 10 Global Importers (Total Value)')
ax2.set_xlabel('Total Import Value (in millions)')
ax2.set_ylabel('Country/Region')
```

EXPORTS
- The United States is the largest individual exporter after "World".
- Other strong exporters include the United Kingdom, Germany, Ireland, India, Netherlands, France, China, and Singapore.
- Most of these are advanced economies, meaning they produce and sell a wide variety of goods globally.
- India also appears in the top exporters, showing its growing role in global trade.

IMPORTS
- The United States is also the biggest importer after "World".
- Ireland, Germany, Netherlands, United Kingdom, France, China, Japan, and Singapore are among the top importers.
- This indicates these countries buy a lot of goods from the rest of the world to meet demand or support industries.

KEY INSIGHTS- 
- The United States is both a leading exporter and importer, showing it’s central in global trade.
- European countries like Germany, Ireland, Netherlands, UK, and France are heavily involved in both exporting and importing.
- China also appears in both lists, reflecting its dual role as a major manufacturing hub and consumer.
- Countries like India and Japan appear only once (India in exports, Japan in imports), which shows their strengths are more on one side of trade.

Understand Economic Indicators & Units of Trade

```python
trade_by_indicator = df.groupby('INDICATOR_LABEL')['OBS_VALUE'].sum().sort_values(ascending=False)

#Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x=trade_by_indicator.values, y=trade_by_indicator.index, palette='crest')
plt.title('Total Trade Value by Economic Indicator')
plt.xlabel('Total Trade Value (in millions)')
plt.ylabel('Indicator')
plt.show()

#Pie Chart
trade_by_unit = df.groupby('UNIT_MEASURE_LABEL')['OBS_VALUE'].sum()

plt.figure(figsize=(8, 8))

colors = sns.color_palette("Set2", len(trade_by_unit))
explode = [0.05] * len(trade_by_unit)

plt.pie(trade_by_unit, 
        labels=None, 
        autopct='%1.1f%%', 
        startangle=140, 
        colors=colors, 
        explode=explode)


plt.legend(trade_by_unit.index, title="Unit of Measure", 
           loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.title('Distribution of Trade Value by Unit of Measure')
plt.show()
```

Bar Chart: Total Trade Value by Economic Indicator
- The chart shows trade values summed up by different economic indicators.
- The largest contributors are:
- International exports in digitally-deliverable services
- International imports in digitally-deliverable services
- These two indicators dominate the chart, meaning digitally-deliverable services (like IT services, software, digital media, cloud services, etc.) 
  account for the bulk of international trade value in your dataset.

Pie Chart: Distribution of Trade Value by Unit of Measure
- The pie chart shows how trade values are measured across units.
- US Dollars make up almost the entire chart (≈99.9%), meaning that most of your dataset’s trade indicators are expressed in monetary terms.
- A very tiny fraction (≈0.1%) is measured in percentages (like “proportion of businesses using computers/Internet”).
- Because of this imbalance, the chart is almost fully covered by the “US Dollars” category, with percentages being almost invisible

Outlier & Distribution Analysis

```python
# Boxplot for export and import values to identify outliers
plt.figure(figsize=(14, 6)) 
sns.boxplot(x='INDICATOR_LABEL', y='OBS_VALUE', data=df)
plt.title('Distribution of Export and Import Values')
plt.xlabel('Trade Type')
plt.ylabel('Trade Value')
plt.ylim(0, 1000000) 


plt.xticks(rotation=45, ha='right')  
plt.tight_layout() 
plt.show()


# Histogram to check the distribution of trade values
plt.figure(figsize=(10, 6))
sns.histplot(df['OBS_VALUE'], bins=50, kde=True)
plt.title('Histogram of Trade Values')
plt.xlabel('Trade Value')
plt.ylabel('Frequency')
plt.yscale('log') 
plt.show()
```

Boxplot- Distribution of Export and Import Values
- The boxplot shows a very wide range of trade values (export & import).
- The median line (inside the box) indicates the typical trade value, but the long whiskers and presence of outliers suggest large fluctuations in    trade.
- Outliers at the top highlight that some trade transactions are extremely high in value compared to the majority, which skews the distribution.
- This implies that while most trades are moderate, a few massive transactions dominate the dataset.

Histogram- Distribution of Trade Values
- The histogram shows the frequency of trade values across ranges.
- The distribution looks highly skewed, with most values concentrated towards the lower end, while very few reach the extreme high values.
- The unusual shape (bars spread unevenly, with large gaps and spikes) suggests:
   The dataset might have sparse values (not continuous trade amounts, but specific transaction bands).
   Some bins have zero counts, creating gaps.
- This confirms what the boxplot showed: most trades are clustered at lower values, but a few very high-value trades stretch the scale.

Trend & Pattern Discovery for Policy and Business

```python
country_of_interest = 'World'

country_df = df[df['REF_AREA_LABEL'] == country_of_interest]
country_trade_over_time = country_df.pivot_table(index='TIME_PERIOD', 
                                                 columns='INDICATOR_LABEL', 
                                                 values='OBS_VALUE', 
                                                 aggfunc='sum')

#line plot
plt.figure(figsize=(12, 6))
country_trade_over_time.plot(ax=plt.gca(), marker='o')
plt.title(f'Trade Trends for {country_of_interest}')
plt.xlabel('Year')
plt.ylabel('Trade Value (in millions)')
plt.legend(title='Trade Type')
plt.grid(True)
plt.show()

#scatter plot
plt.figure(figsize=(8, 8))
sns.scatterplot(data=country_trade_over_time, x='International exports in digitally-deliverable services', y='International imports in digitally-deliverable services')
plt.title(f'Exports vs Imports for {country_of_interest}')
plt.xlabel('Exports')
plt.ylabel('Imports')
plt.grid(True)
plt.show()
```

Line Graph-
- Both exports (blue line) and imports (orange line) show a steady upward trend from 2010 to 2023.
- This reflects the global expansion of digitally-deliverable services (like IT services, digital products, online platforms).
- The gap between exports and imports widens significantly after 2020, showing stronger growth in exports.
- By 2023, exports are around 4.5M, while imports are around 3.75M.
- This shows that digitally-deliverable services are resilient even during disruptions like COVID-19 and have become a major driver of trade.

Scatter Plot-
- The scatter plot shows a positive correlation between exports and imports across countries or regions.
- As exports increase, imports also tend to rise, suggesting that countries engaged in global trade often participate actively on both sides.
- The data points form a roughly linear pattern, indicating a consistent relationship between how much a country sells abroad and how much it buys.
- This pattern reflects how interconnected economies are nations that produce and export more also tend to consume and import more.
